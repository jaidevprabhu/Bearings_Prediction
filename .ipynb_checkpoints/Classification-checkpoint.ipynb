```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import RocCurveDisplay

from sklearn.pipeline import Pipeline
```

```python
results_folder = f'data/results/'
full_data = 'full_data_with_failure_marking.csv'
```

```python
df = pd.read_csv(
    results_folder + full_data,
    index_col=False,
    header = 0,
)
```

```python
df.tail()
```

### Data Cleaning and pre-processing

```python
# Features to consider
columns =  ['min', 'max', 'std', 'kurtosis', 'rms', 'skew', 'shannon_entropy']
```

```python
df[columns].head()
```

```python
df.info()
```

```python
df[df.isna().any(axis=1)]
```

Looks like 5 values for the stats have a NaN, since that's a small proportion of the ~46K values, we can drop them.

```python
# Drop any NaNs
df.dropna(inplace = True)
```

### Target Variable

`class = 0` is negative class - i.e. no problem in the machinery.

`class = 1` is the positive class - i.e. problem is detected.

```python
df['class'].value_counts(normalize=True)
```

We note that we have a mismatched class - we will review later whether to pursue different strategies to resolve it.

```python
# Features that are relevant to classification - we ignore those related to time, & test artifacts such as 
#   id, channel and test #
features_to_scale = df.drop(columns=['time', 'id', 'test', 'channel']).columns.tolist()  
```

```python
X = df[features_to_scale].drop(columns = ['class'])
```

```python
y = df['class']
```

```python
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, 
    y,
    test_size=0.25,
    shuffle=True,
    random_state = 123,
)
```

### Classifier `LogisticRegression`

```python
estimator = LogisticRegression()
```

```python
# Create the pipeline
pipeline = Pipeline(steps=[
    ('scalar', StandardScaler()),
    ('classifier', estimator)
])
```

```python
# Now you can fit the pipeline on your data
pipeline.fit(X_train, y_train)
```

```python
y_preds = pipeline.predict(X_test)
```

```python
rep = classification_report(y_test, y_preds)
```

```python
print(rep)
```

```python
ConfusionMatrixDisplay.from_predictions(
    y_test,
    y_preds,
)
```

```python
RocCurveDisplay.from_estimator(pipeline, X_test, y_test, name = 'LogisticRegression')
```

### Classifier - `SVM`

```python
estimator = SVC()
```

```python
# Create the pipeline
pipeline = Pipeline(steps=[
    ('scalar', StandardScaler()),
    ('classifier', estimator)
])
```

```python
# Now you can fit the pipeline on your data
pipeline.fit(X_train, y_train)
```

```python
y_preds = pipeline.predict(X_test)
```

```python
rep =classification_report(y_test, y_preds)
```

```python
print(rep)
```

```python
svc_cm_display = ConfusionMatrixDisplay.from_predictions(
    y_test,
    y_preds,
)

plt.savefig('images/CM_SVM.png');
```

```python
ROC_SMV_Display = RocCurveDisplay.from_estimator(
    pipeline, 
    X_test, 
    y_test, 
    name = 'SVM'
)

plt.savefig('images/ROC_SVM.png');
```
