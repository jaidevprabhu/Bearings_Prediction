```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
results_folder = f'../data/results/'
file_name_full_data = 'full_data_with_failure_and_pre-fault_marking.csv'
```

```python
df = pd.read_csv(
    results_folder + file_name_full_data,
    index_col=False,
    header = 0,
)
```

```python
stats_features =  ['min', 'max', 'std', 'kurtosis', 'rms', 'skew', 'shannon_entropy']
```

```python
df[stats_features].head()
```

### Corrleation Matrix

```python
correlation_matrix = df[['class'] + stats_features].corr()

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.savefig('../images/corr_matrix_heatmap.png')
plt.show()
```

### Correlation via Pairplot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a pairplot
plt.figure(figsize=(16, 16))
sns.pairplot(df, hue='class', vars=stats_features)
plt.title('Pairplot of Features Colored by Class')
#plt.savefig('../images/corr_pairplot.png')
plt.show()
```
